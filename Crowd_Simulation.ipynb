```python
import pygame
import random
import math

pygame.init()

szerokosc_ekranu, wysokosc_ekranu = 800, 600
screen = pygame.display.set_mode((szerokosc_ekranu, wysokosc_ekranu))
pygame.display.set_caption("Symulacja tlumu")
bialy = (255, 255, 255)
czarny = (0, 0, 0)
szary = (100, 100, 100)

xcel, ycel = 400, 300
xcel2 = 800

class Agent:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.speed = random.uniform(0.75, 1)
        self.cel = 0
        self.slow_down_duration = 0
        self.slow_down_factor = 1.0
        self.vy = 1
        self.vx = 1

    def ruch(self):
        if self.slow_down_duration > 0:
            self.slow_down_duration -= 1
            if self.slow_down_duration == 0:
                self.slow_down_factor = 1.0
                self.vx *= 0.9
                self.vy *= 0.9
            self.x += self.vx * self.slow_down_factor
            self.y += self.vy * self.slow_down_factor

        if self.x == xcel and self.y == ycel:
            self.cel = 1

        if self.cel == 0:
            self.cx = xcel - self.x
            self.cy = ycel - self.y
            dystans_do_celu = math.sqrt(self.cx**2 + self.cy**2)
            proporcja_do_x = self.cx / dystans_do_celu
            proporcja_do_y = self.cy / dystans_do_celu

            self.vx = self.speed * proporcja_do_x
            self.vy = self.speed * proporcja_do_y

        self.x += self.vx
        self.y += self.vy
        self.zawrotka()

    def zawrotka(self):
        if self.x - 8 < 0 or self.x + 8 > szerokosc_ekranu:
            self.vx *= -1
        if self.y - 8 < 0 or self.y + 8 > wysokosc_ekranu:
            self.vy *= -1

    def czy_odbij(self, other_agent):
        dx = self.x - other_agent.x
        dy = self.y - other_agent.y
        dystans = math.sqrt(dx**2 + dy**2)
        return dystans < 12

    def rozodb(self, other_agent):
        if self.czy_odbij(other_agent):
            dx = self.x - other_agent.x
            dy = self.y - other_agent.y
            dystans = math.sqrt(dx**2 + dy**2)
            if dystans == 0:
                dystans = 0.001
            overlap = 6 - dystans
            slow_down_factor = 0.7
            slow_down_time = 40

            nx = (dx / dystans) / 3
            ny = (dy / dystans) / 3

            self.vx -= overlap * 0.01 * nx
            self.vy -= overlap * 0.01 * ny
            other_agent.vx += overlap * 0.01 * nx
            other_agent.vy += overlap * 0.01 * ny

            max_speed = 1
            self.vx = max(-max_speed, min(self.vx, max_speed))
            self.vy = max(-max_speed, min(self.vy, max_speed))
            other_agent.vx = max(-max_speed, min(other_agent.vx, max_speed))
            other_agent.vy = max(-max_speed, min(other_agent.vy, max_speed))

            self.slow_down_factor = slow_down_factor
            self.slow_down_duration = slow_down_time
            other_agent.slow_down_factor = slow_down_factor
            other_agent.slow_down_duration = slow_down_time

    def cien(self):
        pygame.draw.circle(screen, szary, (int(self.x) + 2, int(self.y) + 2), 8)
        pygame.draw.circle(screen, czarny, (int(self.x), int(self.y)), 8)

agenci = []
for i in range(100):
    agenci.append(Agent(random.randint(9, szerokosc_ekranu - 9), random.randint(9, wysokosc_ekranu - 9)))

running = 1
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = 0
    screen.fill(bialy)

    for agent in agenci:
        agent.ruch()
        agent.cien()
        for other_agent in agenci:
            if agent != other_agent:
                agent.rozodb(other_agent)
    pygame.display.flip()
    pygame.time.delay(15)

pygame.quit()
```
